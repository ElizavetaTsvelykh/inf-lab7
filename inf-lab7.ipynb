```python

class Student:
    def __init__(self, name):
        self.name = name
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def calculate_average_grade(self):
        total_sum = sum(self.grades)
        return total_sum / len(self.grades) if len(self.grades) > 0 else 0

    def __repr__(self):
        return f"{self.name}: {self.calculate_average_grade()}"

def main():
    students = []
    students_with_avg_grade_min_4 = []

    students.append(Student("Иван"))
    students.append(Student("Петр"))
    students.append(Student("Анна"))

    students[0].add_grade(4)
    students[1].add_grade(5)
    students[2].add_grade(2)

    for student in students:
        avg_grade = student.calculate_average_grade()
        if avg_grade >= 4:
            students_with_avg_grade_min_4.append(student)

    print("Список студентов со средним баллом не менее 4:")
    students_with_avg_grade_min_4.sort(key=lambda x: -x.calculate_average_grade())
    for student in students_with_avg_grade_min_4:
        print(f"{student.name}: {student.calculate_average_grade():.2f}")

if __name__ == "__main__":
    main()
```
